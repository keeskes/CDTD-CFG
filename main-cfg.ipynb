# Example: Continuous Diffusion for Mixed-Type Tabular Data on the adult dataset

Below, we briefly showcase the application of our CDTD model to the adult dataset. This repository simplifies the use of CDTD in practice by providing a wrapper around the training and sampling processes. Note that the hyperparameter choices for the score model differ slightly from those in the paper. In the paper, we ensure that all benchmark models have roughly the same parameters on the adult dataset.

Note that to increase training and sampling speed further, you could use `torch.compile` or, if you have access to a capable GPU, the TensorFloat32 type via `torch.set_float32_matmul_precision('high')`.

```python
print("Notebook started...")

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import torch
from sklearn.preprocessing import OrdinalEncoder, QuantileTransformer, StandardScaler

from cdtd_cfg import CDTD

print(torch.__version__)
```

```python
df = pd.read_csv("data/adult.csv")

cat_features = [
    "income",
    "workclass",
    "education",
    "marital-status",
    "occupation",
    "relationship",
    "race",
    "gender",
    "native-country",
]
cont_features = [feat for feat in df.columns if feat not in cat_features]

# specify integer-valued continuous features (to be rounded after generation)
int_features = ["fnlwgt", "capital-gain", "capital-loss", "age", "educational-num", "hours-per-week"]

# code missings
df = df.replace(r" ", np.nan)
df = df.replace(r"?", np.nan)
df = df.replace(r"", np.nan)

# remove rows with missings in targets
idx_target_nan = df["income"].isna().to_numpy().nonzero()[0]
df.drop(labels=idx_target_nan, axis=0, inplace=True)

# for categorical features, replace missings with 'empty
df[cat_features] = df[cat_features].fillna("empty")

# # for continuous data, drop missings
df.dropna(inplace=True)

# # ensure correct types
X_cat = df[cat_features].to_numpy().astype("str")
X_cont = df[cont_features].to_numpy().astype("float")

# # preprocess categorical classes and convert to integers
cat_int_enc = OrdinalEncoder()
X_cat_enc = cat_int_enc.fit_transform(X_cat)

# # preprocess continuous features using quantile transformation
cont_enc = QuantileTransformer(
    output_distribution="normal",
    n_quantiles=max(min(X_cont.shape[0] // 30, 1000), 10),
    subsample=int(1e9),
    random_state=42,
)
X_cont_enc = cont_enc.fit_transform(X_cont)
cont_std = StandardScaler()
X_cont_enc_std = cont_std.fit_transform(X_cont_enc)

# ensure correct formats
X_cat_train = torch.tensor(X_cat_enc).long()
X_cont_train = torch.tensor(X_cont_enc_std).float()
```

```python
# Define whether to use cfg and the class features per columns
cfg = True
label_1 = "income"
label_2 = "gender"

label_1_idx = cat_features.index(label_1)
label_2_idx = cat_features.index(label_2)

# Extract the tensor of all the values of these features within the training set
y_condition_1 = X_cat_train[:, label_1_idx] 
y_condition_2 = X_cat_train[:, label_2_idx]  

# initialize and train model
cdtd = CDTD(
    X_cat_train,
    X_cont_train,
    cfg = cfg,
    y_condition_1=y_condition_1,
    y_condition_2=y_condition_2,
)

# Since I have a CPU only PyTorch, there seems no real reason to add this line (it might speed up the process 10%, but nothing drastic)
# cdtd.diff_model = torch.compile(cdtd.diff_model)

# We define the proportion of the model trained unconditionally. 
dropout_ratio=0.2

cdtd.fit(X_cat_train, X_cont_train, num_steps_train=10000, num_steps_warmup=2000, cfg = cfg, 
         y_condition_1=y_condition_1, y_condition_2=y_condition_2, 
         dropout_ratio = dropout_ratio)
```

```python
# If we are performing CFG, we decide if we only want to create data for a single class-combination, or all possible combinations
targeted_sampling = True
# If we perform targeted sampling we select the target values for both labels 
target_label_1 = "<=50K"
target_label_2 = "Male"
cfg_scale = 0.6 # How much we weigh the conditional score functions during the sampling process in total

if cfg:
    label_1_idx = cat_features.index(label_1)
    label_2_idx = cat_features.index(label_2)

    # Get the mapping of encoded classes
    classes_label_1 = cat_int_enc.categories_[label_1_idx]
    classes_label_2 = cat_int_enc.categories_[label_2_idx]

    # Defines the number of categories in both label feature
    num_classes_1 = len(classes_label_1)
    num_classes_2 = len(classes_label_2)

    # Creates the probabilities for each category in the label features
    probs_label_1 = np.zeros(num_classes_1)
    probs_label_2 = np.zeros(num_classes_2)

    if targeted_sampling:
        # We map target string to index, in order to set the target indexes probability to 1
        target_label_1_idx = np.where(classes_label_1 == target_label_1)[0][0]
        target_label_2_idx = np.where(classes_label_2 == target_label_2)[0][0]

        probs_label_1[target_label_1_idx] = 1.0
        probs_label_2[target_label_2_idx] = 1.0
        

    else:
        # Without targetted sampling, we retrieve the frequency of each category from the training data
        y_label_1_train = X_cat_train[:, label_1_idx].numpy()
        y_label_2_train = X_cat_train[:, label_2_idx].numpy()

        for c in range(num_classes_1):
            probs_label_1[c] = np.mean(y_label_1_train == c)
        for c in range(num_classes_2):
            probs_label_2[c] = np.mean(y_label_2_train == c)

    probs_label_1 = probs_label_1 / probs_label_1.sum() # Standardizing to make sure they add up to 1.0
    probs_label_2 = probs_label_2 / probs_label_2.sum()

    
# sample data
num_samples = round(X_cat_train.shape[0])

if cfg:
    X_cat_gen, X_cont_gen = cdtd.sample(num_samples=num_samples, cfg=cfg, 
                                        probs_label_1=probs_label_1, probs_label_2=probs_label_2, cfg_scale = cfg_scale)
    
else: 
    X_cat_gen, X_cont_gen = cdtd.sample(num_samples=num_samples)

# post-process data
X_cat_gen = cat_int_enc.inverse_transform(X_cat_gen)
X_cont_gen = cont_std.inverse_transform(X_cont_gen)
X_cont_gen = cont_enc.inverse_transform(X_cont_gen)

# combine to synthetic dataset
df_gen = pd.concat((pd.DataFrame(X_cat_gen), pd.DataFrame(X_cont_gen)), axis=1)
df_gen.columns = cat_features + cont_features

# round integer features
df_gen[int_features] = df_gen[int_features].round().astype("int")
```

```python
# some functions to simplify plotting

def plot_cat_dist(df_true, df_gen, label, ax):
    counts_true = pd.DataFrame(df_true[label]).value_counts()
    props_true = counts_true / sum(counts_true)
    counts_gen = pd.DataFrame(df_gen[label]).value_counts()
    props_gen = counts_gen / sum(counts_gen)

    # account for categories possibly missed by model
    x = []
    y_true_list = []
    y_gen_list = []

    for cat in props_true.keys():
        x.append(cat[0])
        y_true_list.append(props_true[cat])
        if cat in props_gen.keys():
            y_gen_list.append(props_gen[cat])
        else:
            y_gen_list.append(0.0)

    ax.bar(
        np.arange(len(x)) - 0.1,
        y_true_list,
        color="tab:orange",
        width=0.2,
        label="true",
    )
    ax.bar(
        np.arange(len(x)) + 0.1,
        y_gen_list,
        color="tab:blue",
        width=0.2,
        label="generated",
    )
    ax.set_title(label)
    ax.set_xticks(np.arange(len(x)), x)
    ax.set_xticklabels(x, rotation=90, ha='center', fontsize=8)

    return fig

def plot_cont_dist(df_true, df_gen, label, ax):
    sns.kdeplot(df_true[label], ax=ax, color="tab:orange")
    sns.kdeplot(df_gen[label], ax=ax, color="tab:blue")
    plt.legend(["true", "generated"])
    return fig


# plot categorical data
labels = ["income", "workclass", "education", "occupation", "relationship", "gender"]

fig, ax = plt.subplots(2, 3, figsize=(12, 6))
for i, a in enumerate(ax.flatten()):
    plot_cat_dist(df, df_gen, labels[i], ax=a)
fig.tight_layout(h_pad=1, w_pad=0.4)
plt.legend(["true", "generated"])
```

```python
# plot continuous data
labels = cont_features
fig, ax = plt.subplots(2, 3, figsize=(12, 6))
for i, a in enumerate(ax.flatten()):
    plot_cont_dist(df, df_gen, labels[i], ax=a)
fig.tight_layout(h_pad=1, w_pad=0.4)
```
